````python
from springtime.datasets.insitu.pep725 import PEP725Phenor

dataset = PEP725Phenor(species="Syringa vulgaris")
dataset.download()
df_obs = dataset.load()
df_obs
````

````python
import pandas as pd

pd.read_csv(dataset._location)
````

````python
from springtime.utils import NamedArea

area = NamedArea(
    **{
        "name": "Germany",
        "bbox": [
            5.98865807458,
            47.3024876979,
            7.0169958839,
            49.983104153,
        ],
    }
)


PEP725Phenor(species="Syringa vulgaris").load(
    extract_area=area,
    extract_years=[2000, 2002],
    set_index=["year", "geometry"],
    include_cols=["day"],
)
````

````python
dataset = (
    PEP725Phenor()
    .species("Syringa vulgaris")
    .extract_area(area)
    .extract_years(2000, 2002)
    .include_cols(["day"])
    .set_index(["year", "geometry"])
)
dataset.download()
dataset.load()
dataset.to_recipe()
````

```yaml
datasets:
  observations:
    provider: PEP725Phenor
    provider_args:
      species: Syringa vulgaris
    onload:
      extract_area:
        name: Germany
        bbox: [5.98865807458, 47.3024876979, 7.0169958839, 49.983104153]
      filter_years: [2000, 2002]
      extract_index: true
      include_cols: ["day", "species", "bbch"]
```
